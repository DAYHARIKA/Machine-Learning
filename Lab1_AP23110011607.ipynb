## <a name='reshape'></a>5Ô∏è‚É£ Reshaping, Transpose & Copies vs‚ÄØViews

## <a name='broadcast'></a>8Ô∏è‚É£ Broadcasting Rules

Rules: compare dimensions from right ‚Üí left; stretch size 1 dims; mismatch error.

```python
row = np.arange(5)
col = np.arange(3).reshape(3,1)
matrix = row + col  # broadcast to 3√ó5
print(matrix)
```

```python
# üñäÔ∏è TODO: use broadcasting to create a 10√ó10 multiplication table.

# Create row and column vectors
row = np.arange(1, 11).reshape(1, 10)  # shape (1, 10)
col = np.arange(1, 11).reshape(10, 1)  # shape (10, 1)

# Use broadcasting to multiply
multiplication_table = col * row

print("10√ó10 Multiplication Table:\n", multiplication_table)
```

## <a name='structured'></a>1Ô∏è‚É£1Ô∏è‚É£ Structured / Record Arrays

```python
people = np.array([('Alice', 25, 55.0), ('Bob', 30, 85.5)],
                   dtype=[('name','U10'), ('age','i4'), ('weight','f4')])
print(people['name'], people['age'].mean())
```

```python
# üñäÔ∏è TODO: add a new field 'height' to the structured array using `np.lib.recfunctions.append_fields` (hint: pip install?).
import numpy as np
from numpy.lib import recfunctions as rfn

# Original structured array
people = np.array([('Alice', 25, 55.0), ('Bob', 30, 85.5)],
                  dtype=[('name','U10'), ('age','i4'), ('weight','f4')])

print("Original:")
print(people['name'], people['age'].mean())

# Heights to add (must match shape)
heights = np.array([165.0, 175.5], dtype='f4')

# ‚úÖ Add 'height' field
people_with_height = rfn.append_fields(people, 'height', heights, usemask=False)

print("\nAfter adding 'height' field:")
print(people_with_height)
```

## <a name='mask'></a>1Ô∏è‚É£5Ô∏è‚É£ Masked Arrays & NaNs

```python
arr = np.array([1, 2, np.nan, 4, np.nan])
masked = np.ma.masked_invalid(arr)
print(masked.mean())
```

```python
# üñäÔ∏è TODO: replace NaNs with column means in a 2‚ÄëD array containing NaNs.
# Sample 2D array with NaNs
arr = np.array([
    [1.0, 2.0, np.nan],
    [4.0, np.nan, 6.0],
    [7.0, 8.0, 9.0]
])

# ‚úÖ Compute column means ignoring NaNs
col_means = np.nanmean(arr, axis=0)

# ‚úÖ Find indices where NaNs occur
inds = np.where(np.isnan(arr))

# ‚úÖ Replace NaNs with corresponding column means
arr[inds] = np.take(col_means, inds[1])

print("Array after replacing NaNs with column means:")
print(arr)
```

```python
fitness = np.genfromtxt('fitness.txt', delimiter='\t', dtype=None, encoding=None, names=True)
print('columns:', fitness.dtype.names, 'rows:', len(fitness))
```

```python

# Convert to DataFrame
df = pd.DataFrame(fitness)

# Normalize column names to lowercase
df.columns = df.columns.str.strip().str.lower()

# Convert 'date' column to datetime with dayfirst format
df['date'] = pd.to_datetime(df['date'], dayfirst=True)

# Convert numeric columns to appropriate float types
df['step_count'] = pd.to_numeric(df['step_count'], errors='coerce')
df['mood'] = pd.to_numeric(df['mood'], errors='coerce')
df['calories_burned'] = pd.to_numeric(df['calories_burned'], errors='coerce')
df['hours_of_sleep'] = pd.to_numeric(df['hours_of_sleep'], errors='coerce')
df['weight_kg'] = pd.to_numeric(df['weight_kg'], errors='coerce')

# üü¢ 1. Monthly step count
monthly_steps = df.groupby(df['date'].dt.to_period('M'))['step_count'].sum()
print("Monthly Step Count:\n", monthly_steps)

# üü¢ 2. Sleep vs Mood correlation
correlation = df['hours_of_sleep'].corr(df['mood'])
print("\nSleep vs Mood Correlation:", round(correlation, 3))

# üü¢ 3. Weekly summary of calories burned
weekly_calories = df.groupby(df['date'].dt.to_period('W'))['calories_burned'].sum()
print("\nWeekly Calories Burned:\n", weekly_calories)

```
